## Matrices

On our previous discussions, we mainly dwell in 2-dimensional vectors in which in real-life there is more to that. Today, we'll tackle on multi-dimensional data/vectors and how to deal with them.

Take for example these three vectors:

$$X = \begin{bmatrix} 1\\7 \\\end{bmatrix} , Y = \begin{bmatrix} 3\\10 \\\end{bmatrix} $$

These vectors can be considered as three different sets unrelated of lines. But usually in real-life problems we want to associate or group lines depending on what they represent, or different aspects of a system representing interrelated linear equations. 

We can take a look at the linear equations of these vectors:

$$X = x + 7y \\ Y = 3x + 10y$$

Now if we say that these two lines are interrelated or if we want to find a relationship between them, then we can say they are part of the same "system". Then we would need to declare them as a single vector instead.

$$A = \begin{bmatrix} 1 \ 7\\3 \ 10 \\\end{bmatrix}\\ or \\
A = \left\{
    \begin{array}\\
        x + 7y \\ 
        3x + 10y
    \end{array}
\right. $$

Here we transposed vector $X$ and $Y$ from a column vector to a row. Now that we have a matrix that represents two linear equations, it is now what we call a system of linear equations.<br>

## Subspaces

Now that we have declared a 2-equation vector we can safely call it a space. And the former $X$ and $Y$ vector are subspaces of our new space, specifically $X$ and $Y$ are now <b>row spaces</b>.<br>
Alternately, the <b>column space</b> of the vector will be the x and y values $\begin{bmatrix} 1\\3 \\\end{bmatrix} $ and $\begin{bmatrix} 7\\10 \\\end{bmatrix} $ respectively. <br>
To plot this we do it by:

```python
A = np.array([
    [1, 7],
    [3, 10]
])

plt.xlim(0,15)
plt.ylim(0,15)

# plt.quiver(0,0, A[0,0], A[0,1], angles='xy', scale_units='xy',scale=1, color='red') 
# plt.quiver(0,0, A[1,0], A[1,1], angles='xy', scale_units='xy',scale=1, color='blue')


plt.quiver([0,0],[0,0], A[:,0], A[:,1], 
           angles='xy', scale_units='xy',scale=1, 
           color=['red','blue'])## use column spaces
plt.show()
```

```python
## Continuous line plotting

A = np.array([
    [1, 7],
    [3, 10]
])

c = np.arange(0,100)
plt.xlim(0,15)
plt.ylim(0,15)

plt.plot(c*A[0,0], c*A[0,1], color='red') 
plt.plot(c*A[1,0], c*A[1,1], color='blue') 


plt.show()
```

## 3D Vectors

Now that you have a grasp on 2D vectors, we can now try to visualize in 3D! The idea of 3-Dimensional vectors is that we are now considering 3 different parameters instead of two. In this sense, as we increase the number of considerations or parameters in a system we can theoretically have until infinite number of dimensions of a vector.

Let's take this 3D vector equation:

$$V3 = 3x + 2y + 8z$$

Here we have three different parameters x,y, and z. This also reflects that we have 3 distinct unit vectors we can linearly transform. <br>
Let's try to visualize this:

```python
V3 = np.array([3,2,8])

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim([0, 8])
ax.set_ylim([0, 8])
ax.set_zlim([0, 8])

ax.quiver(0, 0, 0, V3[0], V3[1], V3[2], arrow_length_ratio=0.05, colors='r')

plt.show()
```

```python
V3 = np.array([3,2,8])
fig = plt.figure()
ax = fig.gca(projection='3d')
c = np.arange(0,8)
ax.set_xlim([0, 8])
ax.set_ylim([0, 8])
ax.set_zlim([0, 8])

ax.plot(c*V3[0],c*V3[1],c*V3[2], color='g')
plt.show()
```

Now let's try to visualize a three-dimensional system of linear equation. Take this system for example:

$$
G = \left\{
    \begin{array}\\
        x + 2y + 3z \\ 
        x + y \\
        y + 5z
    \end{array}
\right. $$

```python
G = np.array([
    [1,2,3],
    [1,1,0],
    [0,1,5]
])

fig = plt.figure()
ax1 = fig.gca(projection='3d')

ax1.set_xlim([0, 5])
ax1.set_ylim([0, 5])
ax1.set_zlim([0, 5])

origin = (0,0,0)
ax1.quiver(origin, origin, origin, G[:,0], G[:,1], G[:,2], 
           arrow_length_ratio=0.05, colors=['red','blue','green'])
plt.show()
```

```python
G = np.array([
    [1,5,3],
    [3,1,0],
    [0,1,5]
])
fig = plt.figure()
ax1 = fig.gca(projection='3d')

c = np.arange(0,3)

ax1.set_xlim([0, 5])
ax1.set_ylim([0, 5])
ax1.set_zlim([0, 5])


colors = ['g','r','b']
for i in range(G.shape[0]):
    ax1.plot(c*G[i,0],c*G[i,1],c*G[i,2], color=colors[i])
plt.show()
```

## Reshaping

There are times that you would be required to transform a vector/ matrix into specified dimensions or shapes. NumPy provides the necessary functions for that throug the <b> reshape() </b> function.

Let's say you have a matrix of 3D vectors $
A = \begin{bmatrix} 
1 \ 7 \ 3 \\
3 \ 10\ 2 \\
2 \ 7\ 4 \\
1 \ 2\ 2 \\
\end{bmatrix}$ and you want to turn it from 3D vectors to 2D vectors retaining the values you would do it as:

```python
A = np.array([
    [1, 7, 3],
    [3, 10, 2],
    [2, 7, 4],
    [1, 2, 2]
])
print(A.shape)
print(A.ndim)
print(A.size)
```

```python
A_reshaped = A.reshape((6,2)) ## take note that you can only reshape by factors of the length of the vector
print(A_reshaped)
print(A_reshaped.shape)
print(A_reshaped.ndim)
print(A_reshaped.size)
```

```python
A_reshaped = A.reshape((1,12)) ## take note that you can only reshape by factors of the length of the vector
print(A_reshaped)
print(A_reshaped.shape)
print(A_reshaped.ndim)
print(A_reshaped.size)
```
